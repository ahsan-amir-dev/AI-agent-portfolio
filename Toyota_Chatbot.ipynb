```python
# Project 2: Toyota Dealership Chatbot with Memory
# Built using Python and Groq Cloud API (Llama 3.1 8B)
# Acts as a showroom assistant tracking context for car sales and appointment bookings.

from groq import Groq
client = Groq(api_key="YOUR_GROQ_API_KEY_HERE")

Business_info = """
You are a customer support agent for car showroom name "Toyota Motors".
Act kind and friendly towards customers.
Opening hours is 8am to 6pm daily and closed on sunday.
All models of toyota cars are available on our showroom.
If customer want to know price of cars search on internet and tell them according to 2026.
Book the car for the customers if they want to.
Welcome customers with a friendly greeting.
"""

conversation = [
    {"role":"system", "content": Business_info}
]
def ask_ai(user_message):
  conversation.append({"role": "user", "content": user_message})
  response = client.chat.completions.create(
      model = "llama-3.1-8b-instant",
      messages = conversation
  )
  reply = response.choices[0].message.content
  conversation.append({"role": "assistant", "content": reply})
  return reply
  print("===WELCOME TO TOYOTA SIALKOT CITY MOTORS OR PRESS EXIT FOR QUIR===")

while True:
  user_input = input("customer: ")
  if user_input.lower() == "exit":
     print("THANKS FOR COMING")
     break

  answer = ask_ai(user_input)
  print(f"agent: {answer}")
```
