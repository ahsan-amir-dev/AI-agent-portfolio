```python
# Project 1: AHSAN SALOON Customer Support Chatbot
# Built using Python and Groq Cloud API (Llama 3.1 8B)
# Handles automated customer interactions, styling trends, and salon details.

from groq import Groq
client = Groq(api_key="YOUR_GROQ_API_KEY_HERE")

business_info = """
You are a customer support agent for "AHSAN SALOON" saloon.
Opening hour: 9am to 9pm daily.
We can cut any style and type of hair and beard.
We accept only cash.
Suggest customer trending haircut style.
Keep answer short and friendly.
"""

conversation = [
    {"role": "system", "content": business_info}
]

def ask_ai(user_message):
    conversation.append({"role": "user", "content": user_message})
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=conversation
    )
    reply = response.choices[0].message.content
    conversation.append({"role": "assistant", "content": reply})
    return reply

print("===WELCOME TO AHSAN SALOON=== Type 'exit' to quit.")
while True:
    question = input("Customer: ")
    if question.lower() == "exit":
        print("Goodbye!")
        break
    answer = ask_ai(question)
    print(f"Agent: {answer}")
```
